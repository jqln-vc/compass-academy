```python
import pandas as pd
from ingestao import concatenador_csv
dataset = concatenador_csv('./dados_raw',
                           'dataset.csv',
                           delimitador=';',
                           encoding='utf-8')
df = pd.read_csv('dataset.csv', delimiter=';', encoding='utf-8')
```

```python
df.head(100)
```

```python
colunas = ['Título Original',
           'CPB',
           'Data de Emissão do CPB',
           'Situação da Obra',
           'Tipo da Obra',
           'Subtipo da Obra',
           'Classificação da Obra',
           'Organização Temporal',
           'Duração Total (Minutos)',
           'Quantidade de Episódios',
           'Ano de Produção Inicial',
           'Ano de Produção Final',
           'Segmento de Destinação',
           'Coprodução Internacional',
           'Requerente',
           'CNPJ do Requerente',
           'UF do Requerente',
           'Município do Requerente']
```

```python
df.columns = colunas
df.head(20)
```

```python
df['Tipo da Obra'] = df['Tipo da Obra'].str.capitalize()
df['Tipo da Obra']
```

```python
df['SUBTIPO_OBRA'].unique()
```

```python
df.info()
```

* FILTRAR DADOS: MÍNIMO 2 OPERADORES LÓGICOS
* 2 FUNÇÕES DE AGREGAÇÃO
* 1 FUNÇÃO CONDICIONAL
* 1 FUNÇÃO CONVERSÃO
* 1 FUNÇÃO DE DATA
* 1 FUNÇÃO DE STRING

* CONVERSÃO (FLOAT -> INT)

```python
df['ANO_PRODUCAO_INICIAL'] = df['ANO_PRODUCAO_INICIAL'].fillna(0).astype(int)
df['ANO_PRODUCAO_FINAL'] = df['ANO_PRODUCAO_FINAL'].fillna(0).astype(int)
```

* DATA

```python
df['DATA_EMISSAO_CPB'] = pd.to_datetime(df['DATA_EMISSAO_CPB'], dayfirst=True, errors='coerce')
df['DATA_EMISSAO_CPB']
```

* STRING + REPLACE + CONVERSÃO (FLOAT)

```python
df['DURACAO_TOTAL_MINUTOS'] = df['DURACAO_TOTAL_MINUTOS'].str.replace(',', '.').astype(float)
df['DURACAO_TOTAL_MINUTOS']
```

* FUNÇÃO DE DATA + 2 OPERADORES LÓGICOS

```python
df = df[(df['DATA_EMISSAO_CPB'].dt.year >= 2020) & 
   (df['DATA_EMISSAO_CPB'].dt.year <= 2022) & 
   (df['ORGANIZACAO_TEMPORAL'] == 'NÃO SERIADA')]
```

```python
df.groupby([
    'TIPO_OBRA',
    'SUBTIPO_OBRA',
    'SEGMENTO_DESTINACAO_INICIAL'
])['CPB'].count().reset_index(name='QTD_FILMES').sort_values(by='QTD_FILMES', ascending=False)
```
